```python
#New columns as max of other columns
#float to int used 

df=pd.read_csv('tips.csv')
print(df.shape)
df['sum']=df.sum(axis=1)
df['min']=df.min(axis=1)
df['max']=df.max(axis=1)
df['average']=df.mean(axis=1).astype(int)
print(df)

```

```python
def math_special(row):
    return (row.maths*2+row.language/2+row.history/3+row.science)/4

df=pd.read_csv('abc.csv')
print(df.shape)
df['sum']=df.sum(axis=1)
df['min']=df.min(axis=1)
df['max']=df.max(axis=1)
df['average']=df.mean(axis=1).astype(int)
df['math_special']=df.apply(math_special,axis=1).astype(int)
print(df)
```

```python
#Task 25:
#New columns as max of other columns
#35 marks considered as pass
#If the student fails in math, consider fail
#If the student passes in language and science, consider as pass 
def pass_one_subject(row):
    if(row.maths>34):
        return 'pass'
    if(row.language>34 and row.science>34):
        return 'pass'
    return 'fail'

df=pd.read_csv('abc.csv')
df['pass_one']=df.apply(pass_one_subject,axis=1)

print(df)
```

```python
#Expanding Apply
import functools
series=pd.Series([i/100.0 for i in range(1,6)])
print(series)
def CumRet(x,y):
    return x*(1+y)
def Red(x):
    return functools.reduce(CumRet,x,1.0)
s2=series.expanding().apply(Red)
print(s2)
    
```
