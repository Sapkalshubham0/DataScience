```python
df['GST']=df['total_bill'].apply(lambda total_bill:total_bill*0.10)
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
df['Tip_Quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
df.head()
```

```python
len(df[df['Tip_Quality']=='Generous'])
```

```python
len(df[df['Tip_Quality']=='Other'])
```

```python
df['Tip_Quality']=np.vectorize(quality)(df['total_bill'],df['tip'])
```
