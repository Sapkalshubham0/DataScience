```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tips.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
plt.figure(figsize=(12,8))
sns.displot(df['total_bill'])
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(data=df,x='tip',y='total_bill',hue='smoker',style='sex',palette='Set1')
```

```python
sns.jointplot(data=df,x='tip',y='total_bill',kind='kde') #kind of plot to draw
```

```python
sns.barplot(x='sex',y='price_per_person',data=df)
```

```python
sns.boxplot(data=df,x='day',y='total_bill',hue='smoker')
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex')
```
